```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import warnings
warnings.simplefilter("ignore")
%matplotlib inline
```

```python
df = pd.read_csv("avocado.csv")
df = df.drop(df.columns[0], axis=1)
```

```python
df.head()
```

`Date` - The date of the observation

`AveragePrice` - the average price of a single avocado

`type` - conventional or organic

`year` - the year

`Region` - the city or region of the observation

`Total Volume` - Total number of avocados sold

`4046` - Total number of avocados with PLU 4046 sold

`4225` - Total number of avocados with PLU 4225 sold

`4770` - Total number of avocados with PLU 4770 sold

```python
df["type"].value_counts()
df["Date"] = pd.to_datetime(df.Date)
```

```python
df.groupby("region").get_group("TotalUS")[['Date', 'AveragePrice']].reset_index(drop=True)\
.plot(x='Date', y='AveragePrice',kind = "line",figsize = (15,10))
plt.xticks(rotation=60)
plt.show();
```

```python
plt.figure(figsize=(12,5))
plt.title("Distribution Price")
ax = sns.distplot(df["AveragePrice"],fit=norm)
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(y="type", x="AveragePrice", data=df);
```
